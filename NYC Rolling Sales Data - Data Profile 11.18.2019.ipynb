NYC Rolling Sales Data last 12 months (October 2018-October 2019) <br>
Data Dictionary: https://www1.nyc.gov/assets/finance/downloads/pdf/07pdf/glossary_rsf071607.pdf

```python
import pandas as pd
```

```python
# Test the import of data from NYC Rolling Sales Data by importing
# the Manhattan data
# skip the first 4 rows as they contain information about the file
manhattan = pd.read_excel('https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_manhattan.xls', skiprows = 4)
```

```python
# see the first 5 rows of the Manhattan dataframe
manhattan.head()
```

```python
# Import the excel data for each of the other 4 boroughs into Pandas Dataframes
bronx = pd.read_excel('https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_bronx.xls', skiprows = 4)
brooklyn = pd.read_excel('https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_brooklyn.xls', skiprows = 4)
queens = pd.read_excel('https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_queens.xls', skiprows = 4)
statenisland = pd.read_excel('https://www1.nyc.gov/assets/finance/downloads/pdf/rolling_sales/rollingsales_statenisland.xls', skiprows = 4)
```

```python
#Combine all the dataframes for each borough into one dataframe 
#for all of NYC 
nyc = pd.concat([manhattan,bronx,brooklyn,queens,statenisland])
```

```python
# See how many listings there are for each borough
nyc['BOROUGH'].value_counts()
```

```python
# Get summary of the numerical information
# Notice that borough is included as it is considered numerical (needs to be changed to actual values)
nyc.describe()
```
